```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ptb_words.csv")
epoch = range(data.shape[0])
plt.plot(epoch, data.loss, 'b-', epoch, data.val_loss, 'r-')
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(["training loss", "validation loss"])
plt.title("Word level LSTM learning curve (loss)")
plt.savefig("ptb_word_loss.jpg")
plt.show()

tp = np.exp(data.loss)
vp = np.exp(data.val_loss)

plt.plot(epoch, tp, 'b-', epoch, vp, 'r-')
plt.ylabel("perplexity")
plt.xlabel("epoch")
plt.legend(["training perplexity", "valalidation perplexity"])
plt.title("Word level LSTM learning curve (perplexity)")
plt.savefig("ptb_word_perplexity.jpg")
plt.show()
```
